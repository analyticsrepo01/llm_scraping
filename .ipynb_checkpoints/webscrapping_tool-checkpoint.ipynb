```python
from crewai_tools import ScrapeWebsiteTool

# To enable scrapping any website it finds during it's execution
tool = ScrapeWebsiteTool()

url = 'https://cloud.google.com/blog/products/ai-machine-learning/rags-powered-by-google-search-technology-part-1#:~:text=Beyond%20our%20expertise,left%20by%20each'

url = "https://seekingalpha.com/article/4702202-delta-air-lines-stock-ultimate-contrarian-buy?source=first_level_url%3Alatest-articles%7Ccontent_type%3Aall%7Csection%3Acontent%7Cline%3A1"
# Initialize the tool with the website URL, so the agent can only scrap the content of the specified website
tool = ScrapeWebsiteTool(website_url=url)

# Extract the text from the site
text = tool.run()
print(text)
```
