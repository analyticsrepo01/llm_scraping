```python
from crewai_tools import SeleniumScrapingTool

# Example 1: Initialize the tool without any parameters to scrape the current page it navigates to
tool = SeleniumScrapingTool()


website_url = "https://github.com/OpenDevin/OpenDevin"
# Example 2: Scrape the entire webpage of a given URL
tool1 = SeleniumScrapingTool(website_url='https://github.com/OpenDevin/OpenDevin')

# Example 3: Target and scrape a specific CSS element from a webpage
tool2 = SeleniumScrapingTool(website_url=website_url, css_element='.main-content')

# Example 4: Perform scraping with additional parameters for a customized experience
tool3 = SeleniumScrapingTool(website_url=website_url, css_element='.main-content', cookie={'name': 'user', 'value': 'John Doe'}, wait_time=10)
```

```python
print(tool1,'\n',tool2,'\n','\n',tool3)
```
